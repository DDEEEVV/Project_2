```python
#importing libraries and dependencies 

import yfinance as yf
import pandas as pd 
import numpy as np 
from pathlib import Path
```

```python
#using yahoo finance API to generate data for Bitcoin
cryptocurrencies = ['BTC-USD']
start_date = "2020-01-01"
end_date = "2022-07-11"
bitcoin_data = yf.download(cryptocurrencies, start_date, end_date)
```

```python
#keeping data only for the first period of the analysis:

btc_1st_year = bitcoin_data.loc["2020-07-01":"2021-07-31"]
btc_1st_year
```

```python
from ta import add_all_ta_features
from ta.utils import dropna
```

```python
pd.set_option('display.max_columns', None)

btc_1st_year = dropna(btc_1st_year)
btc_1st_year = add_all_ta_features(
    btc_1st_year, open="Open", high="High", low="Low", close="Close", volume="Volume")

btc_1st_year = btc_1st_year.fillna(0)
btc_1st_year.head(2)
```

```python
#reading data from the result of sentiment analysis done using tweeter API

btc_sentiments = pd.read_csv(
    Path("Resources/BTC_sentiment_analysis.csv")
)
btc_sentiments.tail(2)
```

```python
#setting the date as index and keeping only the data needed for the analysis

btc_sentiments_1st_year = btc_sentiments.set_index("date")
btc_sentiments_1st_year.index = pd.to_datetime(btc_sentiments_1st_year.index)
btc_sentiments_1st_year = btc_sentiments_1st_year.loc["2020-07-01":"2021-07-31"]
btc_sentiments_1st_year.tail(2)
```

```python
top8_btcholders_covid_cases = pd.read_csv(
    Path("Resources/top8_btc_covid_cases.csv")
)

covid_cases_1st_year = top8_btcholders_covid_cases.set_index("date")
covid_cases_1st_year.index = pd.to_datetime(covid_cases_1st_year.index)
covid_cases_1st_year = covid_cases_1st_year.loc["2020-07-01":"2021-07-31"]
covid_cases_1st_year.tail(2)
```

```python
covid_cases_1st_year = covid_cases_1st_year["Total_cases"].pct_change()
covid_cases_1st_year = covid_cases_1st_year
covid_cases_1st_year.tail(2)
```

```python
#collecting data from WHO about accumulation covid-19 deaths for the first period
top8_btcholders_covid_deaths = pd.read_csv(
    Path("Resources/top8_btc_covid_deaths.csv")
)

covid_deaths_1st_year = top8_btcholders_covid_deaths.set_index("date")
covid_deaths_1st_year.index = pd.to_datetime(covid_deaths_1st_year.index)
covid_deaths_1st_year = covid_deaths_1st_year.loc["2020-07-01":"2021-07-31"]
covid_deaths_1st_year.tail(2)
```

```python
covid_deaths_1st_year = covid_deaths_1st_year["Total_deaths"].pct_change()
covid_deaths_1st_year = covid_deaths_1st_year
covid_deaths_1st_year.tail(2)
```

```python
#contacting bitcoin data with added features, the results of the sentiment analysis and covid cases and death for top 8 countries that has the most bitcoin:
 
btc_analysis_1st_year = pd.concat([btc_1st_year, btc_sentiments_1st_year, covid_cases_1st_year, covid_deaths_1st_year], axis=1, join="inner")
btc_analysis_1st_year = btc_analysis_1st_year.drop(columns=["text"])
btc_analysis_1st_year.index.name = "Date"
display(btc_analysis_1st_year.head(2))
display(btc_analysis_1st_year.tail(2))
```

```python
btc_analysis_1st_year.plot(figsize=(16,8))
```

```python
y = btc_analysis_1st_year["Close"].pct_change().dropna()
y = (y > 0)
y.value_counts()
y = y.loc["2020-07-05":"2021-07-31"]
y
```

```python
X = btc_analysis_1st_year.shift(periods=2).dropna()
X = X.loc["2020-07-01":"2021-07-30"]
X
```

```python
X.shape,y.shape
```

```python
num_test = 33
X_train = X[:-num_test]
y_train = y[:-num_test]
X_test = X[-num_test:]
y_test = y[-num_test:]
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=4, random_state=17)
clf.fit(X_train, y_train)
```

```python
res = clf.predict(X_test)
acc = (res== y_test.values).mean()
print(acc)
```

```python
%matplotlib inline
import  matplotlib.pyplot as plt

colors = ['green' if x else 'red' for x in res== y_test]

ax = plt.gca()
ax.plot(X_test['Close'],lw = 1, c = 'black')
X_test.reset_index().plot.scatter( x = 'Date', y = 'Close',color = colors,s = 48,ax= ax,figsize=(18,9))
ax.xaxis.set_major_locator(plt.MaxNLocator(4))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, res))
```
