```python
import yfinance as yf
import pandas as pd 
import numpy as np
```

```python
cryptocurrencies = ['BTC-USD']
start_date = "2020-07-01"
end_date = "2021-07-31"
bitcoin_data = yf.download(cryptocurrencies, start_date, end_date)
```

```python
bitcoin_data
```

```python
from ta import add_all_ta_features
from ta.utils import dropna
```

```python
pd.set_option('display.max_columns', None)

bitcoin_data = dropna(bitcoin_data)
bitcoin_data = add_all_ta_features(
    bitcoin_data, open="Open", high="High", low="Low", close="Close", volume="Volume")

bitcoin_data = bitcoin_data.fillna(0)
bitcoin_data
```

```python
bitcoin_data = bitcoin_data.drop(columns=["volatility_bbl","volatility_bbhi", "volatility_bbli", "volatility_kchi" , "trend_psar_up_indicator", "trend_psar_down_indicator", "momentum_ppo"])
bitcoin_data
```

```python
y = bitcoin_data["Close"].pct_change().dropna()
y = (y > 0)
y.value_counts()
y = y.loc["2020-07-04":"2021-07-31"]
y
```

```python
X = bitcoin_data.shift(periods=2).dropna()
X = X.loc["2020-07-02":"2021-07-30"]
X
```

```python
X.shape,y.shape
```

```python
num_test = 33
X_train = X[:-num_test]
y_train = y[:-num_test]
X_test = X[-num_test:]
y_test = y[-num_test:]
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=4, random_state=17)
clf.fit(X_train, y_train)
```

```python
# work on this and keep only features that are useful for the model

features = pd.DataFrame({"gini": clf.feature_importances_, "features":bitcoin_data.columns})
features.head()
```

```python
features.plot(kind="bar", figsize = (20,5))
```

```python
features.describe()
```

```python
res = clf.predict(X_test)
acc = (res== y_test.values).mean()
print(acc)
```

```python
%matplotlib inline
import  matplotlib.pyplot as plt

colors = ['green' if x else 'red' for x in res== y_test]

ax = plt.gca()
ax.plot(X_test['Close'],lw = 1, c = 'black')
X_test.reset_index().plot.scatter( x = 'Date', y = 'Close',color = colors,s = 48,ax= ax,figsize=(18,9))
ax.xaxis.set_major_locator(plt.MaxNLocator(4))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, res))
```
